```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()

url = "https://www.kia.com/kr/vehicles/kia-ev/guide/faq"
driver.get(url)

time.sleep(2)
```

```python
faq_buttons = driver.find_elements(By.CSS_SELECTOR, "[id$='-button']")
print("FAQ 버튼 개수:", len(faq_buttons))
```

```python
data = []

faq_buttons = driver.find_elements(By.CSS_SELECTOR, "[id$='-button']")

for i in range(len(faq_buttons)):
    # 매 반복마다 다시 찾기
    faq_buttons = driver.find_elements(By.CSS_SELECTOR, "[id$='-button']")
    button = faq_buttons[i]
    
    # 질문 텍스트
    question = button.text.strip()
    
    # 버튼 id에서 번호 추출
    button_id = button.get_attribute("id")   # 예: accordion-item-0-button
    panel_id = button_id.replace("-button", "-panel")
    
    # 화면 중앙으로 이동
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", button)
    time.sleep(1)
    
    # 클릭
    driver.execute_script("arguments[0].click();", button)
    time.sleep(2)
    
    # 답변 추출
    answer_selector = f"#{panel_id} > div > div"
    
    try:
        answer = driver.find_element(By.CSS_SELECTOR, answer_selector).text.strip()
    except:
        answer = ""
    
    data.append({
        "question": question,
        "answer": answer
    })

print("수집 완료:", len(data))
```

```python
for row in data[:5]:
    print("Q:", row["question"])
    print("A:", row["answer"])
    print("-" * 80)
```

```python
df = pd.DataFrame(data)
df.to_csv("kia_ev_faq.csv", index=False, encoding="utf-8-sig")

print("CSV 저장 완료: kia_ev_faq.csv")
```
